```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'top_200_youtubers.csv'
data = pd.read_csv(file_path)

followers_by_category = data.groupby('Category')['followers'].sum().reset_index()
print("Total followers by category:")
print(followers_by_category)

plt.figure(figsize=(12, 6))
plt.barh(followers_by_category['Category'], followers_by_category['followers'], color='skyblue')
plt.xlabel('Number of Followers')
plt.title('Total Followers by Category')
plt.show()

top_channels = data.nlargest(10, 'followers')
print("\nTop 10 channels by followers:")
print(top_channels[['Channel Name', 'followers']])

plt.figure(figsize=(12, 6))
plt.barh(top_channels['Channel Name'], top_channels['followers'], color='orange')
plt.xlabel('Number of Followers')
plt.title('Top 10 Channels by Followers')
plt.show()

engagement_by_country = data.groupby('Country')['Engagement Rate'].mean().reset_index()
print("\nAverage engagement rate by country:")
print(engagement_by_country)

plt.figure(figsize=(12, 6))
plt.barh(engagement_by_country['Country'], engagement_by_country['Engagement Rate'], color='lightgreen')
plt.xlabel('Average Engagement Rate (%)')
plt.title('Average Engagement Rate by Country')
plt.show()

likes_views_by_category = data.groupby('Category')[['Likes', 'Views']].mean().reset_index()
print("\nAverage likes and views by category:")
print(likes_views_by_category)

plt.figure(figsize=(12, 6))
plt.bar(likes_views_by_category['Category'], likes_views_by_category['Likes'], color='purple', alpha=0.6, label='Average Likes')
plt.bar(likes_views_by_category['Category'], likes_views_by_category['Views'], color='cyan', alpha=0.6, label='Average Views')
plt.ylabel('Count')
plt.title('Average Likes and Views by Category')
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Additional analysis: Distribution of Followers
plt.figure(figsize=(12, 6))
plt.hist(data['followers'], bins=30, color='teal', alpha=0.7)
plt.xlabel('Number of Followers')
plt.ylabel('Frequency')
plt.title('Distribution of Followers Among Channels')
plt.show()

# Additional analysis: Correlation between Likes and Views
plt.figure(figsize=(12, 6))
plt.scatter(data['Likes'], data['Views'], alpha=0.6, color='magenta')
plt.xlabel('Likes')
plt.ylabel('Views')
plt.title('Correlation Between Likes and Views')
plt.grid(True)
plt.show()
```
